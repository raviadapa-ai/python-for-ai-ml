```python
# LOAD DATASETS
import pandas as pd
import numpy as np

# ── Load all three datasets ────────────────────────────────────────────
df_metrics = pd.read_csv(
    "server_metrics.csv",
    parse_dates=["timestamp"],
    dtype={"server_id": "category"},
    na_values=["N/A", "null", "—"]
)

df_tickets = pd.read_csv(
    "incidents.csv",
    parse_dates=["created_at", "resolved_at"]
)

df_logs = pd.read_csv(
    "app_logs.csv",
    parse_dates=["timestamp"]
)

# ── Confirm shapes ─────────────────────────────────────────────────────
print(f"df_metrics : {df_metrics.shape}")
print(f"df_tickets : {df_tickets.shape}")
print(f"df_logs    : {df_logs.shape}")
```

```python
print(df_metrics.index)
print(df_metrics.head(3))
```

```python
# Set_index() and reset_index()
```

```python
# Set server_id as index
df_metrics_idx = df_metrics.set_index("server_id")
print(df_metrics_idx.head())
print(df_metrics_idx.index)
```

```python
# Set timestamp as index — more useful for time series
df_metrics_ts = df_metrics.set_index("timestamp")
print(df_metrics_ts.head())
```

```python
# Reset back to default integer index
df_metrics_reset = df_metrics_ts.reset_index()
print(df_metrics_reset.head())
print(df_metrics_reset.index)
```

```python
# loc — label based (use index values)
df_metrics_ts = df_metrics.set_index("timestamp")

# Get all rows for a specific date
print(df_metrics_ts.loc["2026-01-01"])
```

```python
# MultiIndex
```

```python
# Set server_id + timestamp as MultiIndex
df_multi = df_metrics.set_index(["server_id", "timestamp"])
print(df_multi.head(10))
print(df_multi.index)
```

```python
# Slice by outer index — all rows for srv-01
print(df_multi.loc["srv-01"])
```

```python
# Slice by outer + inner index — srv-01 on specific date
df_multi = df_metrics.set_index(["server_id", "timestamp"]).sort_index()

print(df_multi.loc[("srv-01", "2026-01-01")])

# Now slice cleanly — no warning
print(df_multi.loc["srv-01"])
print(df_multi.loc[("srv-01", "2026-01-01")])
```

```python
# sort_index()
# Sort by index ascending (default)
df_metrics_ts = df_metrics.set_index("timestamp").sort_index()
print(df_metrics_ts.head())

# Sort descending — latest readings first
print(df_metrics_ts.sort_index(ascending=False).head())
```

```python
# Boolean filter
boolean_result = df_metrics[
    (df_metrics["server_id"] == "srv-01") &
    (df_metrics["timestamp"] >= "2026-01-01") &
    (df_metrics["timestamp"] <= "2026-01-03 23:59:59")
]
print(f"Boolean filter rows: {len(boolean_result)}")

# Index-based
df_ts = df_metrics.set_index("timestamp").sort_index()
index_result = df_ts.loc["2026-01-01":"2026-01-03"]
index_result = index_result[index_result["server_id"] == "srv-01"]
print(f"Index filter rows: {len(index_result)}")
```

```python
# AIOps Pattern: Timestamp as Index
```

```python
# Set timestamp as index for time-based slicing
df_metrics_ts = df_metrics.set_index("timestamp").sort_index()

# Get all readings between two timestamps
print(df_metrics_ts.loc["2026-01-01":"2026-01-03", ["server_id", "cpu_pct", "status"]])

# Get critical readings in a time window
print(df_metrics_ts.loc["2026-01-01":"2026-01-03"][
    df_metrics_ts.loc["2026-01-01":"2026-01-03"]["status"] == "critical"
])
```
