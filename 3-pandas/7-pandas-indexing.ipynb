```python
# LOAD DATASETS
import pandas as pd
import numpy as np

# ── Load all three datasets ────────────────────────────────────────────
df_metrics = pd.read_csv(
    "server_metrics.csv",
    parse_dates=["timestamp"],
    dtype={"server_id": "category"},
    na_values=["N/A", "null", "—"]
)

df_tickets = pd.read_csv(
    "incidents.csv",
    parse_dates=["created_at", "resolved_at"]
)

df_logs = pd.read_csv(
    "app_logs.csv",
    parse_dates=["timestamp"]
)

# ── Confirm shapes ─────────────────────────────────────────────────────
print(f"df_metrics : {df_metrics.shape}")
print(f"df_tickets : {df_tickets.shape}")
print(f"df_logs    : {df_logs.shape}")
```

```python
# sort_index()
# Sort by index ascending (default)
df_metrics_ts = df_metrics.set_index("timestamp").sort_index()
print(df_metrics_ts.head())

# Sort descending — latest readings first
print(df_metrics_ts.sort_index(ascending=False).head())
```

```python
# Boolean filter
boolean_result = df_metrics[
    (df_metrics["server_id"] == "srv-01") &
    (df_metrics["timestamp"] >= "2026-01-01") &
    (df_metrics["timestamp"] <= "2026-01-03 23:59:59")
]
print(f"Boolean filter rows: {len(boolean_result)}")

# Index-based
df_ts = df_metrics.set_index("timestamp").sort_index()
index_result = df_ts.loc["2026-01-01":"2026-01-03"]
index_result = index_result[index_result["server_id"] == "srv-01"]
print(f"Index filter rows: {len(index_result)}")
```

```python
# MultiIndex datetime slicing is unreliable in pandas
# Production pattern: boolean filter first → then set_index
# Never rely on MultiIndex loc[] slicing for datetime inner level
# MultiIndex with datetime inner level has consistent issues across pandas versions. Senior engineers avoid it and use boolean filtering instead.
```

```python
# Task 7 — cross-dataset
# Using boolean filter on df_metrics, retrieve all critical readings for srv-04. Count them. Then do the same using index-based approach. Confirm both counts match.
```

```python
# Boolean
count_bool = len(df_metrics[
    (df_metrics["server_id"] == "srv-04") &
    (df_metrics["status"] == "CRITICAL")
])

# Index
df_multi = df_metrics.set_index(["server_id", "status"]).sort_index()

try:
    count_index = len(df_multi.loc[("srv-04", "CRITICAL")])
except KeyError:
    count_index = 0

print(count_bool, count_index, count_bool == count_index)
```
