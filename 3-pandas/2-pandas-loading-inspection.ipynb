```python
# Loading data, write to postgreSQL and read back required data through sql query in pandas. No CSV export needed.

import pandas as pd
from sqlalchemy import create_engine

df = pd.read_csv("server_metrics.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])

engine = create_engine("postgresql://postgres:postgres@localhost:5432/postgres")

df.to_sql("server_metrics", engine, if_exists="replace", index=False)
#if_exists="replace" drops and recreates the table every run 

df_output = pd.read_sql("""
    SELECT server_id, timestamp, cpu_pct, memory_pct, response_ms
    FROM server_metrics
    WHERE timestamp >= (
        SELECT MAX(timestamp) - INTERVAL '7 days'
        FROM server_metrics
    )
""", engine)
print(df_output.dtypes)
print(f"shape : {df_output.shape}")
print(f"Date range : {df_output['timestamp'].min()} to {df_output['timestamp'].max()}")
print(f"Row count : {len(df_output)}")
print(f"Max CPU : {df_output['cpu_pct'].max()}")
print(df_output.head())
```

```python
# Task: Load incidents.csv with correct parse_dates. Find:
# Total number of tickets
# Number of unresolved tickets
# Which category appears most
# Date range of tickets (created_at min and max)

import pandas as pd

# check column names

df_tickets = pd.read_csv("incidents.csv")
print(df_tickets.columns.tolist())

# Load data
df_tickets = pd.read_csv("incidents.csv", parse_dates=["created_at"])

# 1. Total number of tickets
total_tickets = len(df_tickets)

# 2. Number of unresolved tickets
unresolved_tickets = df_tickets["resolved_at"].isna().sum()

# 3. Most common category
most_common_category = df_tickets["category"].value_counts().idxmax()

# 4. Date range (min and max of created_at)
min_date = df_tickets["created_at"].min()
max_date = df_tickets["created_at"].max()

# Print results
print("Total tickets:", total_tickets)
print("Unresolved tickets:", unresolved_tickets)
print("Most common category:", most_common_category)
print("Date range:", min_date, "to", max_date)
```

```python
# Task -  Load app_logs.csv. Find the distribution of log_level values as percentages. Which level appears least?
df_logs = pd.read_csv("app_logs.csv")
print(df_logs)
# Distribution of log_level as percentages
log_dist = df_logs["log_level"].value_counts(normalize=True) * 100

print(log_dist)

# Least frequent log level
least_common = log_dist.idxmin()

print("Least frequent log level:", least_common)
```

```python
# Task:  Across all three DataFrames — df_metrics, df_tickets, df_logs — print a summary table showing: DataFrame name, shape, null count total, and date range
import pandas as pd

summary = []

# Helper function
def get_summary(name, df, date_col=None):
    return {
        "DataFrame": name,
        "Shape": df.shape,
        "Total Nulls": df.isna().sum().sum(),
        "Start Date": df[date_col].min() if date_col else None,
        "End Date": df[date_col].max() if date_col else None
    }

# Add summaries (adjust date columns based on your data)
summary.append(get_summary("df_metrics", df_metrics, "timestamp"))
summary.append(get_summary("df_tickets", df_tickets, "created_at"))
summary.append(get_summary("df_logs", df_logs, "timestamp"))

# Convert to DataFrame
df_summary = pd.DataFrame(summary)

print(df_summary)
```
