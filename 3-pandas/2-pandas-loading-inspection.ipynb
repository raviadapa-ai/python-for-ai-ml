```python
# Read_csv - loading with right options
import pandas as pd
df_metrics = pd.read_csv(
    "server_metrics.csv", 
    parse_dates = ["timestamp"],
    dtype = {"server_id" : "category"},
    na_values = ["N/A", "null", "-"]
)
print("shows exact columns names :", df_metrics.columns.tolist(), "\n")
# head() and tail() - first and last look
print("top 5 rows of df_metrics", df_metrics.head())
print("last 5 rows of df_metrics", df_metrics.tail())
df_metrics.head(10)  #without print gives the data in a table format with frames
```
