```python
tickets = [
    {"id": "TK001", "title": "VPN down", "priority": "High"},
    {"id": "TK002", "title": "Printer offline", "priority": "Low"},
    {"id": "TK003", "title": "Email slow", "priority": "Medium"},
    {"id": "TK004", "title": "Server unreachable", "priority": "High"},
    {"id": "TK005", "title": "Password reset", "priority": "Low"}
]

# print all High priority tickets first
# Then Medium
# Then Low
high_priority = []
med_priority = []
low_priority = []
for ticket in tickets:
    if ticket["priority"] == "High":
        high_priority.append(ticket["id"]) 
    elif ticket["priority"] == "Medium":
        med_priority.append(ticket["id"])
    elif ticket["priority"] == "Low":
        low_priority.append(ticket["id"])

print(f"LIST OF HIGH PRIORITY TICKETS:{','.join(high_priority)}")
print(f"LIST OF MED PRIORITY TICKETS: {','.join(med_priority)}")
print(f"LIST OF LOW PRIORITY TICKETS:'{','.join(low_priority)}")
```

```python
#combine two lists

student_name = ["Raj", "kiran", "Ramesh", "Varsha", "Rajan"]
score = [45, 60, 75, 85, 98]
# combine = zip(student_name, score) # doesnt work
combine = list(zip(student_name, score))  #this works
print(combine)
print(max(combine))
print(max(score))
print(min(score))
average = sum(score) / len(score)
print("average", average)


# Print each student with score
for name, s in combine:
    print(name, "-", s)

# Find topper
topper = max(combine, key=lambda x: x[1])
print("\nTopper:", topper[0], "with", topper[1])

# New concept here — lambda
#lambda x: x[1] just means "look at the second item" which is the score. Think of it as a mini one line function:
# This lambda:
key=lambda x: x[1]

# Is same as writing:
def get_score(x):
    return x[1]
```

```python
# Now Let's Combine All Three!

students = {
    "Raj": 45,
    "kiran": 77,
    "Ramesh": 92,
    "Pawan": 98
}

def class_report():
    print("-----------class report-----------")
    for name, score in students.items():
        print(name, "-", score)
        
    average = sum(students.values())/ len(students)
    print("Average - ", average)
    topper = max (students.items(), key = lambda x:x[1])
    print("Upper:", topper[0], topper[1])
    print("Below average")
    for name, score in students.items():
        if score < average: 
            print(name, "-", score)
        else:
            continue
    
class_report()        
```

```python
# Function that checks server health
#  AI monitoring agent

def check_server_health(hostname, cpu_percent, ram_percent, disk_percent):
    alerts = []
    
    if cpu_percent > 85:
        alerts.append(f"HIGH CPU: {cpu_percent}%")
    if ram_percent > 90:
        alerts.append(f"HIGH RAM: {ram_percent}%")
    if disk_percent > 85:
        alerts.append(f"HIGH DISK: {disk_percent}%")
    
    if alerts:
        print(f"ALERT - {hostname}: {', '.join(alerts)}")
        return "CRITICAL"
    else:
        print(f"OK - {hostname}: All systems normal")
        return "HEALTHY"

# Test with different servers
check_server_health("SERVER-01", cpu_percent=45, ram_percent=67, disk_percent=55)
check_server_health("DB-PRIMARY", cpu_percent=91, ram_percent=95, disk_percent=60)
check_server_health("WEB-01", cpu_percent=70, ram_percent=88, disk_percent=87)
```

```python
def check_server_health (host_name, cpu_percent, ram_percent, disk_percent):
    alerts = []
    if cpu_percent > 90:
        alerts.append(f"HIGH CPU : {cpu_percent} %")
    if ram_percent>85:
        alerts.append(f"HIGH RAM USAGE: {ram_percent} %")
    if disk_percent>95: 
        alerts.append(f"DISK FULL: {disk_percent} %")

    if alerts: 
        print(f"ALERT {host_name}: {','.join(alerts)}")
        return "CRITICAL"
    else: 
        print(f"ok - {host_name} : all systems normal")
        return "HEALTHY"
check_server_health("db-01", 45, 55, 65)
check_server_health("db-02", 98, 97, 96)
```

```python
# Now use the function on a list of servers
# production monitoring system

servers = [
    {"hostname": "SERVER-01", "cpu": 45, "ram": 67, "disk": 55},
    {"hostname": "SERVER-02", "cpu": 92, "ram": 88, "disk": 45},
    {"hostname": "DB-PRIMARY", "cpu": 78, "ram": 95, "disk": 60},
    {"hostname": "WEB-01", "cpu": 55, "ram": 45, "disk": 88},
    {"hostname": "APP-01", "cpu": 89, "ram": 91, "disk": 72}
]

print("=== SERVER HEALTH REPORT ===")
critical_servers = []

for server in servers:
    result = check_server_health(
        server["hostname"],
        server["cpu"],
        server["ram"],
        server["disk"]
    )
    if result == "CRITICAL":
        critical_servers.append(server["hostname"])

print(f"\nTotal servers checked: {len(servers)}")
print(f"Critical servers: {len(critical_servers)}")
if critical_servers:
    print(f"Needs attention: {', '.join(critical_servers)}")
```
