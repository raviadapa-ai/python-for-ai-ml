```python
# try/except/else/finally
try:
    with open("server_log.txt", "r") as f:
        content = f.read()

except FileNotFoundError:
    print("Log file missing")

else:
    print("File read successfully")    # runs only if NO error

finally:
    print("Script complete")           # ALWAYS runs — error or not
# # Block    When it runs
# try          Always — the code to attempt
# except       Only when error occurs
# else         Only when NO error occurs
# finally      Always — with or without error
```

```python
#IT Ops example
import json
def load_config(filename):
    try:
        with open(filename, "r") as f:
            config = json.load(f)
            return config
    except FileNoFoundError:
        print("Config file not found:", filename)
        return None
    except json.JSONDecodeErro:
        print("Config file is corrupted or invalid JSON")
        return None
    finally:
        print("Config load attempt complete")

config = load_config("config.json")

if config:
    print("Host", config["host_name"])
else:
    print("Using default settings")
```

```python
# Raising Your Own Exceptions

def check_cpu(value):
    if value < 0 or value > 100:
        raise ValueError("CPU value must be between 0 and 100")
    return value

try:
    check_cpu(150)
except ValueError as e:
    print("Error:", e)
```

```python
# Safe Value Converter
# function called safe_convert that takes a string value like "85%" or "abc%" and tries to convert it to integer after removing %. 
# Handle ValueError if conversion fails. Test with both valid and invalid values.

def safe_convert(string):
    try: 
        value = int(string.replace("%",""))
        return value
    
    except ValueError:
        print("Invalid value — cannot convert to integer")
        return None

result = safe_convert("num%")
print("Safe convert:", result)

result2 = safe_convert("abc%")
print("Safe convert:", result2)

string = input("Enter the percentage % ")
print(string)
type(string)
print("Safe convert", safe_convert(string))
```

```python
# Load config.json, handle missing file and corrupted JSON, print all values if successful, always print "System ready" at the end.

import json
config = {
    "hostname" : "10.0.0.17",
    "database" : "sql",
    "cpu" : 45,
    "disk" : 95
}

with open ("config.json", "w") as f:
    json.dump(config, f, indent=4)
print("config saved")

try:
    with open ("config.json", "r") as f:
        config = json.load(f)
        print("Host:", config["hostname"])
        print("Port:", config["database"])
        print("CPU Threshold:", config["disk"])

except FileNotFoundError:
    print("Config not found — using defaults")
except json.JSONDecodeError:
    print("Config corrupted")
finally:
    print("system ready")
```
