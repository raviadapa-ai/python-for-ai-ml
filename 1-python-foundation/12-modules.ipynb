```python
# Importing Modules
import os
import sys
import datetime

#os - interact with operating system
print(os.getcwd()) #current working directory
print(os.listdir(".")) # list files in current directory

# sys - system information
print(sys.version)  # python version
print(sys.platform) #window/linux/darwin

#datetime
now = datetime.datetime.now()
print(now)
print(now.strftime("%Y-%m-%d"))
```

```python
#importing spectific function instead of importing the whole module

from datetime import datetime
from os import getcwd

now = datetime.now() # no need for datetime.datetime
print(now.strftime("Y-%m-%d %H-%M"))

print(getcwd()) # no need for os.getcwd()
```

```python
# import with alias - different names
import datetime as dt
import os as operating_system
now = dt.datetime.now()
print(operating_system.getcwd())
print(now)
print(now.strftime("%Y-%m-%d"))
```
