```python
# This is a real log analysis script — reads from file, processes each line, reports summary.
#  Reading + Processing Together
#  String Methods + File Handling combine:

critical_count = 0
warning_count = 0
ok_count = 0

with open("server_log1.txt", "r") as f:
    for line in f:
        line = line.strip()
        parts = line.split(" | ")
        server = parts[1]
        metric = parts[2]
        value = int(parts[3].replace("%", ""))

        if value > 85:
            print("CRITICAL —", server, "|", metric, "|", str(value) + "%")
            critical_count += 1
        elif value > 70:
            print("WARNING  —", server, "|", metric, "|", str(value) + "%")
            warning_count += 1
        else:
            print("OK       —", server, "|", metric, "|", str(value) + "%")
            ok_count += 1

print("Total CRITICAL:", critical_count)
print("Total WARNING:", warning_count)
print("Total OK:", ok_count)
```

```python
#  File Log Parser
# Using server_log.txt from Part 1:
# Read line by line
# Print only lines where value is above 80%
# Print total count at the end
#Then write the parser — read the file, filter above 80%, count them

with open("server_log1.txt", "r") as f:
    count = 0
    for line in f:
        line = line.strip()
        parts = line.split(" | ")
        server = parts[1]
        name = parts[2]
        value = int(parts[3].replace("%", ""))
        if value > 80:
            print("CRITICAL", server, name, value)
            count = count + 1
print("No of critical counts", count)    
```
