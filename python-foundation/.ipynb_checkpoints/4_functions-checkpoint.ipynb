```python
# function called `calculate` that takes **two numbers and an operation** and returns the result.

def calculate (a,b,operation):
    if operation == 'add': 
        return a + b
    elif operation == 'subtract':
        return a-b
    elif operation == 'division': 
        return a/b
    elif operation == 'multiply':
        return a*b
a = int(input("enter number a:"))
b = int(input("enter number b:"))

operation = input("enter operation:")
print("calculations are", calculate(a,b,operation))
```

```python
def calculate(a, b, operation):
    if operation == 'add':
        return a + b
    elif operation == 'subtract':
        return a - b
    elif operation == 'multiply':
        return a * b
    elif operation == 'divide':
        return a / b
    else:
        return "Invalid operation"

a = int(input("enter number a: "))
b = int(input("enter number b: "))
operation = input("enter operation: ")

print("Result:", calculate(a, b, operation))
```

```python
def calculate_progress(completed_weeks, total_weeks):
    percentage = (completed_weeks / total_weeks) * 100
    remaining = total_weeks - completed_weeks
    
    print("=== Progress Report ===")
    print("Completed:", completed_weeks, "weeks")
    print("Remaining:", remaining, "weeks")
    print("Progress: {:.1f}%".format(percentage))  #The {:.1f} formats the number to one decimal place.
    
    if percentage < 25:
        print("Status: Early stage — build the habit")
    elif percentage < 50:
        print("Status: Building momentum — stay consistent")
    elif percentage < 75:
        print("Status: Strong progress — specialization phase")
    else:
        print("Status: Final stretch — transition imminent")

# Now USE the function multiple times
calculate_progress(0, 28)
print()
calculate_progress(7, 28)
print()
calculate_progress(14, 28)
print()
calculate_progress(21, 28)
```

```python
#lists and functions together

# find server health status

server1_readings = [45, 52, 48, 61, 55, 49, 58]
server2_readings = [72, 78, 81, 75, 79, 83, 77]
server3_readings = [88, 92, 87, 95, 91, 89, 93]

def assess_readings(reading):
    total = sum(reading)
    average = total / len(reading)
    return average
    
readings = [server1_readings, server2_readings, server3_readings]

for reading in readings:
    result = assess_readings(reading)
    if result > 95:
        print("CRITICAL:", result, reading)
    elif result >80:
        print("WARNING", result, reading)
    else:
        print("HEALTHY", result, reading)
```
