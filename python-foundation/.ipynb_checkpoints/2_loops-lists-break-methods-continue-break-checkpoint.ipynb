````python
#if,elif conditions in lists
statuses = [
    "online", "offline", "online", "maintenance", 
    "online", "offline", "online", "maintenance", "online"
]

# Your job: write code that prints:
# Online: 5
# Offline: 2  
# Maintenance: 2
a=b=c=0
for status in statuses:
    if status == "online":
       a = a+1
    elif status == "offline":
        b = b+1
    elif status == "maintenance":
        c = c+1
print("online", a)
print("offline", b)
print("maintenance",c)
    
````

````python
#for,elif combined in lists
statuses = [
    "online", "offline", "online", "maintenance",
    "online", "offline", "online", "maintenance", "online"
]

online_count = 0
offline_count = 0

maintenance_count = 0

for status in statuses:
    if status == "online":
        online_count += 1
    elif status == "offline":
        offline_count += 1
    elif status == "maintenance":
        maintenance_count += 1

print(f"Online: {online_count}")
print(f"Offline: {offline_count}")
print(f"Maintenance: {maintenance_count}")
````

````python
server1_readings = [45, 52, 48, 61, 55, 49, 58]
server2_readings = [72, 78, 81, 75, 79, 83, 77]
server3_readings = [88, 92, 87, 95, 91, 89, 93]

# Your job: write a function called assess_server(readings)
# that returns "healthy", "warning", or "critical"
# Then call it for all three servers and print the result

def assess_readings(reading):
    total = sum(reading)
    average = total / len(reading)
    return average
    
readings = [server1_readings, server2_readings, server3_readings]

for reading in readings:
    result = assess_readings(reading)
    if result > 95:
        print("CRITICAL:", result, reading)
    elif result >80:
        print("WARNING", result, reading)
    else:
        print("HEALTHY", result, reading)
````

````python
# Use while when you don't know how many times to loop — you loop until a condition is met.

connection_attempts = 0

while connection_attempts < 3:
    print("Attempting to connect... attempt", connection_attempts + 1)
    connection_attempts = connection_attempts + 1

print("Max attempts reached. Check network.")
````

````python
items = ["SERVER-01", "SERVER-02", "DB-PRIMARY"]

# Join with comma and space
result = ", ".join(items)
print(result)
# Output: SERVER-01, SERVER-02, DB-PRIMARY

# Join with just a dash
result = " - ".join(items)
print(result)
# Output: SERVER-01 - SERVER-02 - DB-PRIMARY

# Join with nothing — no separator
result = "".join(items)
print(result)
# Output: SERVER-01SERVER-02DB-PRIMARY

# Join with a newline — each item on its own line
result = "\n".join(items)
print(result)
# Output:
# SERVER-01
# SERVER-02
# DB-PRIMARY
````

````python

## Now Let's Go One Level Up

# Write a program that:
# - Stores 5 student names and 5 scores in separate lists
# - Prints each student name with their score
# - Prints the name of the student with the highest score
# ```
# Expected output:
# Alice - 85
# Bob - 90
# Charlie - 78
# David - 92
# Eva - 88

# Topper: Bob with 90
students = ["Ravi", "Pawan", "Nishanth", "Kumar"]
scores = [94,77,85,86]

combine = list (zip(students, scores))
print(combine)
topper = max(combine, key=lambda x:x[1])
print(topper)
````
