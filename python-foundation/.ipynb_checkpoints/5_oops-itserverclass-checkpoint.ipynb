```python
# OOP — Object Oriented Programming
# Instead of writing loose functions and variables, you bundle related data and behaviour together into a class. A class is a blueprint. From that blueprint you create objects.
# IT Ops analogy — a server has a name, CPU, memory. It can generate alerts. Instead of managing those as separate variables, you create a Server class 

#IT SERVER CLASS

class ITServer:
     def __init__(self, name, cpu, memory):
         self.name = name
         self.cpu = cpu
         self.memory = memory

# Creating an object from the class:
server1 = ITServer("SERVER01", 95, 64)
print(server1.name)    # SERVER01
print(server1.cpu)     # 95

#Methods
class ITServer:
     def __init__(self, name, cpu, memory):
         self.name = name
         self.cpu = cpu
         self.memory = memory

     def generate_alert(self):
         if self.cpu > 80:
             print(f"ALERT: {self.name} CPU at {self.cpu}%")
         else:
             print(f"{self.name} is healthy")

server1 = ITServer("SERVER01", 95, 64)
server1.generate_alert()   # ALERT: SERVER01 CPU at 95%

server2 = ITServer("SERVER02", 45, 32)
server2.generate_alert()   # SERVER02 is healthy
```

```python
# Now extend it yourself. Add this to your existing code:
# Task 1 — Create a list of 3 servers and loop through calling generate_alert() on each.
# Task 2 — Add a second method called server_info() that prints name, cpu, and memory together in one f-string.
#Methods
class ITServer:
    def __init__(self, name, cpu, memory):
        self.name = name
        self.cpu = cpu
        self.memory = memory

    def generate_alert(self):
        if self.cpu > 80:
            print(f"ALERT: {self.name} CPU at {self.cpu}%")
        else:
            print(f"{self.name} is healthy")
    def server_info(self):
        print(f"{self.name}, {self.cpu}, {self.memory}")
    
      
server1 = ITServer("SERVER01", 95, 64)
server2 = ITServer("SERVER02", 45, 32)
server3 = ITServer("SERVER03", 48, 98)

servers = [server1, server2, server3]

for server in servers:
    server.server_info()
    server.generate_alert()
```

```python
# Inheritance
# One more step to complete OOP — Inheritance. Write a DatabaseServer class that:

# Inherits from ITServer
# Adds one attribute: db_name
# Adds one method: db_status() that prints the server name and database name
# Create one object and call both generate_alert() and db_status() on it

class ITServer:
    def __init__(self, name, cpu, memory):
        self.name = name
        self.cpu = cpu
        self.memory = memory

    def generate_alert(self):
        if self.cpu > 80:
            print(f"ALERT: {self.name} CPU at {self.cpu}%")
        else:
            print(f"{self.name} is healthy")
    def server_info(self):
        print(f"{self.name}, {self.cpu}, {self.memory}")

class Databaseserver(ITServer):
    def __init__(self, name, cpu, memory,db_name):
        super().__init__(name,cpu,memory)
        self.db_name = db_name

    def db_status(self):
        print(f"{self.name}, {self.db_name}")
      
server1 = ITServer("SERVER01", 95, 64)
server2 = ITServer("SERVER02", 45, 32)
server3 = ITServer("SERVER03", 48, 98)

servers = [server1, server2, server3]

for server in servers:
    server.server_info()
    server.generate_alert()

db1 = Databaseserver("DBServer01", 85, 128, "Postsql")
db1.generate_alert()
db1.db_status()
```
