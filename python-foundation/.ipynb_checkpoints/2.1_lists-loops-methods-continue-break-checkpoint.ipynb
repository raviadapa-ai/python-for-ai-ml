```python
#if,elif, for conditions in lists
# finding how many online, offline and maintenance tickets from the list.

statuses = [
    "online", "offline", "online", "maintenance", 
    "online", "offline", "online", "maintenance", "online"
]

a=b=c=0
for status in statuses:
    if status == "online":
       a = a+1
    elif status == "offline":
        b = b+1
    elif status == "maintenance":
        c = c+1
print("online", a)
print("offline", b)
print("maintenance",c)
    
```

```python
# finding the status of servers whether they are online /offline

servers = ["SERVER-01", "SERVER-02", "DB-PRIMARY", "WEB-01", "APP-01"]
offline = ["SERVER-02", "WEB-01"]

for server in servers:
    if server in offline:
        print(f"OFFLINE: {server}")
    else:
        print(f"ONLINE:  {server}")
    
```

```python
servers = ["server-01", "server-02", "server-03", "db-primary", "web-01", "app-01"]
offline = ["server-02", "web-01"]

for server in servers:
    if server in offline:
        print(f"server is:offline: {server}")
    else:
        print(f"server is online: {server}")
```

```python
#for,elif combined in lists
#counting servers status
statuses = [
    "online", "offline", "online", "maintenance",
    "online", "offline", "online", "maintenance", "online"
]

online_count = 0
offline_count = 0

maintenance_count = 0

for status in statuses:
    if status == "online":
        online_count += 1
    elif status == "offline":
        offline_count += 1
    elif status == "maintenance":
        maintenance_count += 1

print(f"Online: {online_count}")
print(f"Offline: {offline_count}")
print(f"Maintenance: {maintenance_count}")
```

```python
#for loop in lists
weekly_topics = [
    "Python basics",
    "Pandas for data",
    "SQL fundamentals",
    "Statistics",
    "Machine Learning",
    "NLP foundations",
    "LLM applications",
    "RAG systems",
    "AI Agents"
]

print(" complete learning roadmap:")
print("================================")

for topic in weekly_topics:
    print("→", topic)

print("================================")
print("Total topics to master:", len(weekly_topics))
```

```python
#methods in lists  #len
print("complete learning roadmap:")
print("============================")
for topic in weekly_topics: 
    print("->", topic)
print("============================")
print("total topics to master:", len(weekly_topics))
```

```python
#Appending/copying/removing in lists

completed = []
pending = weekly_topics.copy()

completed.append(pending[0])
pending.remove(pending[0])

print("Completed:", completed)
print("Remaining:", len(pending), "topics")
print("Progress:", len(completed), "out of", len(weekly_topics))
```

```python
#join

items = ["SERVER-01", "SERVER-02", "DB-PRIMARY"]

# Join with comma and space
result = ", ".join(items)
print(result)
# Output: SERVER-01, SERVER-02, DB-PRIMARY

# Join with just a dash
result = " - ".join(items)
print(result)
# Output: SERVER-01 - SERVER-02 - DB-PRIMARY

# Join with nothing — no separator
result = "".join(items)
print(result)
# Output: SERVER-01SERVER-02DB-PRIMARY

# Join with a newline — each item on its own line
result = "\n".join(items)
print(result)

items = ["SERVER-01", "SERVER-02", "DB-PRIMARY"]

# Join with comma and space
result1 = "\n".join(items)
print(items)
print(result1)
```

```python
#combining lsits #zip, #lambda, #max

students = ["Ravi", "Pawan", "Nishanth", "Kumar"]
scores = [94,77,85,86]

combine = list (zip(students, scores))
print(combine)
topper = max(combine, key=lambda x:x[1])
print(topper)
```

```python
topper = max(combine, key=lambda x:x[1])
print(topper)
print(topper[0], topper[1])
```

```python
#continue:
servers = ["db01", "web01", "MAINTENANCE", "app01", "cache01"]
for server in servers:
    if server == "MAINTENANCE":
        print("SKIP MAINTENANCE")
        continue
    print("Checking server", server)
```

```python
#break
servers = ["db01", "web01", "MAINTENANCE", "app01", "cache01"]
for server in servers:
    if server == "MAINTENANCE":
        print("Server is under maintenance")
        break      
    print("Checking server", server)
```
