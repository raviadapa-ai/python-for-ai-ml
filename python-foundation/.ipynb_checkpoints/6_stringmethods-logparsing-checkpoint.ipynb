```python
# split()
pythonlog = "2026-03-04 | SERVER01 | CPU | 91%"
parts = pythonlog.split(" | ")
print(parts)

# Access each part by index
print(parts[0])     # 2026-03-04
print(parts[1])     # SERVER01
print(parts[2])     # CPU
print(parts[3])     # 91%
# Split breaks a string into a list wherever it finds the separator you give it.
# Split by comma
ip = "192,168,1,1"
octets = ip.split(",")
print(octets)       # ['192', '168', '1', '1']

# Split by space
sentence = "disk usage critical"
words = sentence.split(" ")
print(words)        # ['disk', 'usage', 'critical']
```

```python
#  find() and count()
log = "ERROR: connection failed. ERROR: timeout"

# count() — how many times does it appear
print(log.count("ERROR"))       # 2

# find() — returns position of first match, -1 if not found
print(log.find("ERROR"))        # 0 — found at position 0
print(log.find("WARNING"))      # -1 — not found

```

```python
log = "ERROR: connection failed. ERROR: timeout. WARNING: retry"

# count() — how many times does it appear
print(log.count("ERROR"))       # 2
print(log.count("WARNING"))     # 1

# find() — position of first match, -1 if not found
print(log.find("WARNING"))      # 43
print(log.find("CRITICAL"))     # -1 — not found
```

```python
# Removing % and converting to int

value = "91%"

# Step 1 — remove the % sign
cleaned = value.replace("%", "")
print(cleaned)      # 91  — but still a string!

# Step 2 — convert to integer
number = int(cleaned)
print(number)       # 91  — now a number you can compare

# One line version
number = int(value.replace("%", ""))
print(number + 9)   # 100 — now you can do math on it
```

```python
#  Removing % and Converting to Int
value = "87%"

# Remove % sign
cleaned = value.replace("%", "")
print(cleaned)        # 87 — still a string!

# Convert to integer
number = int(cleaned)
print(number)         # 87 — now a number

# One line version
number = int(value.replace("%", ""))
print(number + 10)    # 97 — can do math now
```

```python
# Removing percentage
disk = "87%"
memory = "92%"
cpu = "45%"

# 1. Remove % from all three
disk1 = disk.replace("%", "")
memory1 = memory.replace("%", "")
cpu1 = cpu.replace("%", "")

# 2. Convert all to integers
disk_num = int(disk1)
memory_num = int(memory1)
cpu_num = int(cpu1)

# 3. Print which ones are above 80
if disk_num > 80:
    print(f"Disk is above 80: {disk_num}%")
if memory_num > 80:
    print(f"Memory is above 80: {memory_num}%")
if cpu_num > 80:
    print(f"CPU is above 80: {cpu_num}%")
```

```python
# Putting it all together in single string
log_line = "2026-03-04 | SERVER01 | CPU | 91%"

# Split into parts
parts = log_line.split(" | ")

# Extract each field
date   = parts[0]
server = parts[1]
metric = parts[2]
value  = int(parts[3].replace("%", ""))

# Apply logic
if value > 85:
    status = "CRITICAL"
elif value > 70:
    status = "WARNING"
else:
    status = "OK"

print(server, "|", metric, "|", str(value) + "%", "|", status)
```

```python
# Part 8 — Putting It All Together in multiple strings (list)
log_lines = [
    "2026-03-05 | SERVER01 | CPU | 91%",
    "2026-03-05 | SERVER02 | DISK | 65%",
    "2026-03-05 | SERVER03 | MEMORY | 88%",
    "2026-03-05 | SERVER04 | CPU | 72%",
    "2026-03-05 | SERVER05 | DISK | 55%"
]

for log in log_lines:
    parts = log.split(" | ")
    server = parts[1]
    metric = parts[2]
    value  = int(parts[3].replace("%", ""))

    if value > 85:
        print("CRITICAL —", server, "|", metric, "|", str(value) + "%")
    elif value > 70:
        print("WARNING  —", server, "|", metric, "|", str(value) + "%")
    else:
        print("OK       —", server, "|", metric, "|", str(value) + "%")
```

```python
log_lines = [
    "2026-03-05 | SERVER01 | CPU | 91%",
    "2026-03-05 | SERVER02 | CPU | 65%",
    "2026-03-05 | SERVER03 | MEMORY | 88%",
    "2026-03-05 | SERVER04 | CPU | 72%",
    "2026-03-05 | SERVER05 | DISK | 55%"
]

a = b = c = 0
for log in log_lines:
    parts = log.split(" | ")
    server = parts[1]
    metric = parts[2]
    value = parts[3].replace("%", "")
    value1 = int(value)
    
    if value1 > 85 : 
        print("CRITICAL", server, "|", metric, "|", str(value1) + "%")
        a = a + 1
    elif value1 > 70 :
        print("WARNING", server, "|", metric, "|", str(value1) + "%")
        b = b +1
    else:
        print("OK", server, "|", metric, "|", str(value1) + "%")
        c = c +1

print("TOTAL CRITICAL:", a)
print("TOTAL WARNING:", b)
print("TOTAL OK:", c)
```
