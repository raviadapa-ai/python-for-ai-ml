```python
# AIOps — Check server health and give alerts

server_name = "SERVER01"
cpu_usage = 92
memory_usage = 78
disk_usage = 95

# Conditional statement
if cpu_usage > 90 and disk_usage > 90:
    print(f"CRITICAL: {server_name} — CPU {cpu_usage}% & Disk {disk_usage}% both critical!")

elif cpu_usage > 80 or memory_usage > 80:
    print(f"WARNING: {server_name} — High resource usage detected!")

elif disk_usage > 70:
    print(f"INFO: {server_name} — Disk usage getting high, monitor closely.")

else:
    print(f"NORMAL: {server_name} — All systems healthy!")
```

```python
# AIOps — Loop through server numbers and check CPU

print(" AIOps Server Health Report")
print("=" * 35)

for server_number in range(1, 6):

    # Simulate CPU reading for each server
    # In real AIOps this reads from actual server
    if server_number == 1:
        server_name = "SERVER01"
        cpu_usage = 95

    elif server_number == 2:
        server_name = "SERVER02"
        cpu_usage = 45

    elif server_number == 3:
        server_name = "SERVER03"
        cpu_usage = 82

    elif server_number == 4:
        server_name = "SERVER04"
        cpu_usage = 60

    elif server_number == 5:
        server_name = "SERVER05"
        cpu_usage = 91

    # Check health
    if cpu_usage > 90:
        status = " CRITICAL"
    elif cpu_usage > 75:
        status = " WARNING"
    else:
        status = " HEALTHY"

    print(f"{status} | {server_name} | CPU: {cpu_usage}%")

print("=" * 35)
print(f" Total servers checked: {server_number}")
```

```python
# Use while when you don't know how many times to loop — you loop until a condition is met.
# attempting to connect servers
connection_attempts = 0

while connection_attempts < 3:
    print("Attempting to connect... attempt", connection_attempts + 1)
    connection_attempts = connection_attempts + 1  #  connection_attempts+=1
print("Max attempts reached. Check network.")
```

```python
# AIOps — Continuously monitor server CPU 
import time
import random

server_name  = "SERVER01"
check_count  = 0
max_checks   = 5
alert_count  = 0
warning_count = 0
healthy_count = 0

print(f" Starting monitoring: {server_name}")
print("=" * 40)

while check_count < max_checks:

    # Simulate live CPU reading
    cpu_usage = random.randint(40, 100)
    check_count += 1

    if cpu_usage > 90:
        alert_count += 1
        print(f" Check #{check_count} | CPU:{cpu_usage}% | CRITICAL!")

    elif cpu_usage > 75:
        warning_count += 1
        print(f" Check #{check_count} | CPU:{cpu_usage}% | WARNING!")

    else:
        healthy_count += 1
        print(f" Check #{check_count} | CPU:{cpu_usage}% | HEALTHY")

    time.sleep(2)

print("=" * 40)
print(f" Total checks  : {check_count}")
print(f" Critical alerts: {alert_count}")
print(f" Warnings      : {warning_count}")
print(f" Healthy checks: {healthy_count}")
```
