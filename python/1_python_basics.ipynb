````python
#Exercise 3: Making Decisions 
hours_practiced_today = float(input("No of hours_practiced_today:"))

if hours_practiced_today >= 1:
    print("Consistency maintained.")
elif hours_practiced_today > 0:
    print("Something is better than nothing. Do more tomorrow.")
else:
    print("No practice today. The foundation weakens.")
````

````python
weekly_hours = int(input("Enter no of hours to practice in a week"))
months_to_sept_2026 = 7
weeks_to_sept_2026 = months_to_sept_2026 * 4  # roughly Sept 2026

total_hours = weekly_hours * weeks_to_sept_2026
python_hours = total_hours * 0.20
sql_hours = total_hours * 0.15
ml_hours = total_hours * 0.35
llm_hours = total_hours * 0.30

print("=== AI Engineering Journey Calculator ===")
print("No of hours to practice everyday", weekly_hours/7, "hours/day")
print("Total months until sept 2026:", months_to_sept_2026, "months")
print("Total weeks unitl sept 2026:", weeks_to_sept_2026, "weeks")
print("Total learning hours until Sept 2026:", total_hours, "hours")
print("Python and Data foundations:", python_hours, "hours")
print("SQL:", sql_hours, "hours")
print("Machine Learning:", ml_hours, "hours")
print("LLM and AI Engineering:", llm_hours, "hours")

if total_hours >= 150:
    print("\nSufficient foundation possible. Stay consistent.")
else:
    print("\nIncrease daily hours if possible.")
````

````python
# Lists and Loops 
weekly_topics = [
    "Python basics",
    "Pandas for data",
    "SQL fundamentals",
    "Statistics",
    "Machine Learning",
    "NLP foundations",
    "LLM applications",
    "RAG systems",
    "AI Agents"
]

print(" complete learning roadmap:")
print("================================")

for topic in weekly_topics:
    print("→", topic)

print("================================")
print("Total topics to master:", len(weekly_topics))
````
