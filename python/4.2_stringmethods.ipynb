````python
# part5 FIND() AND COUNT()
log = "ERROR: connection failed. ERROR: timeout. WARNING: retry"

# count() — how many times does it appear
print(log.count("ERROR"))       # 2
print(log.count("WARNING"))     # 1

# find() — position of first match, -1 if not found
print(log.find("WARNING"))      # 43
print(log.find("CRITICAL"))     # -1 — not found
````

````python
#Part 7 — Case Insensitive Check
log = "WARNING: High Memory Usage on DB-Server-01"

# This fails
print("memory" in log)           # False — capital M

# This works — convert both to lowercase
print("memory" in log.lower())   # True
# ```

# ---

# ### Why This Matters in Real Logs

# Log formats are **inconsistent** in real environments:
# ```
# "ERROR: disk full"
# "Error: Disk Full"
# "error: DISK FULL"
````

````python
log1 = "WARNING: High Memory Usage on DB-Server-01"
log2 = "error: connection timeout on SERVER-02"
log3 = "Error: Disk Full on Prod-Server"

# 1. Check if "memory" exists in log1 — case insensitive
# 2. Check if "error" exists in log2 — case insensitive
# 3. Check if "error" exists in log3 — case insensitive
print("memory" in log1.lower())
print("error" in log2.lower())
print("error" in log3.lower())
````

````python
log_lines = [
    "2026-03-05 | SERVER01 | CPU | 91%",
    "2026-03-05 | SERVER02 | DISK | 65%",
    "2026-03-05 | SERVER03 | MEMORY | 88%",
    "2026-03-05 | SERVER04 | CPU | 72%",
    "2026-03-05 | SERVER05 | DISK | 55%"
]

for log in log_lines:
    parts = log.split(" | ")
    server = parts[1]
    metric = parts[2]
    value  = int(parts[3].replace("%", ""))

    if value > 85:
        print("CRITICAL —", server, "|", metric, "|", str(value) + "%")
    elif value > 70:
        print("WARNING  —", server, "|", metric, "|", str(value) + "%")
    else:
        print("OK       —", server, "|", metric, "|", str(value) + "%")
````

````python
log_lines = [
    "2026-03-05 | SERVER01 | CPU | 91%",
    "2026-03-05 | SERVER02 | CPU | 65%",
    "2026-03-05 | SERVER03 | MEMORY | 88%",
    "2026-03-05 | SERVER04 | CPU | 72%",
    "2026-03-05 | SERVER05 | DISK | 55%"
]

a = b = c = 0
for log in log_lines:
    parts = log.split(" | ")
    server = parts[1]
    metric = parts[2]
    value = parts[3].replace("%", "")
    value1 = int(value)
    
    if value1 > 85 : 
        print("CRITICAL", server, "|", metric, "|", str(value1) + "%")
        a = a + 1
    elif value1 > 70 :
        print("WARNING", server, "|", metric, "|", str(value1) + "%")
        b = b +1
    else:
        print("OK", server, "|", metric, "|", str(value1) + "%")
        c = c +1

print("TOTAL CRITICAL:", a)
print("TOTAL WARNING:", b)
print("TOTAL OK:", c)
    
````
