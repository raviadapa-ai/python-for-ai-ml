```python
import datetime
class Itserver:
    
    def __init__(self, name, cpu, memory):
        self.name = name 
        self.cpu = cpu
        self.memory = memory
        
    def __str__(self):
        return f"{self.name} | CPU {self.cpu} | MEMORY {self.memory}"

    def get_alert(self):
        if (self.cpu > 80) and (self.memory >85):
            return "ALERT : HIGH CPU & MEMORY"
        elif self.cpu > 80: 
            return "ALERT : HIGH CPU "
        elif self.memory > 85:
            return "ALERT : HIGH MEMORY"
        else:
            return "STATUS : HEALTHY"
        
    def to_log(self):
        return {
            "server" : self.name,
            "cpu" : self.cpu,
            "memory" : self.memory,
            "status" : self.get_alert(),
            "timestamp": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")   
        }
server1 = Itserver("server1", 98, 88)
server2 = Itserver("server2", 88, 68)
server3 = Itserver("server3", 65, 55)
server4 = Itserver("server4", 78, 86)
servers = [server1, server2, server3,server4]

logs = []

for server in servers:
    
    print(server)
    status = server.get_alert()
    print(status)
    
    log_entry = server.to_log()
    logs.append(log_entry)
    print("log entry", log_entry)
    print("-"*44)

print(f"------------- LOG REPORT--------------------")

for log in logs:
    print(f"{log['timestamp']} | {log['server']} | CPU: {log['cpu']} | MEM: {log['memory']} | {log['status']}")
```
